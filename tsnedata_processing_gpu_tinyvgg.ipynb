```python
import numpy as np

list_array = []
values = [0]
sum = 0

labels = []

iList = ['ABE', 'ART', 'BAS', 'BLA', 'EBO', 'EOS', 'FGC', 'HAC', 'KSC', 'LYI', 'LYT', 'MMZ', 'MON', 'MYB', 'NGB', 'NGS', 'NIF', 'OTH', 'PEB', 'PLM', 'PMO']
    
for file in readfiles:
    
    parts = file.split("\\")
    

    # Take the last part (the filename), split it by '.', and take the first part
    filename = parts[-1].split(".")[0]
    
    

    # Remove the last character (the '2') from the filename
    result = filename[:3]
    print(result)
    lbl = iList.index(result)
    
    arr = np.loadtxt(file)
    r,c = arr.shape
    sum += r
    for k in range(len(arr)):
        labels.append(lbl)
    
    list_array.append(arr)
    values.append(sum)




features = np.concatenate(list_array,axis=0)
print(features.shape)
print(values)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score


# load the iris dataset as an example
data = features
print(data)

# split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.3, random_state=0)

# instantiate the model
gnb = MultinomialNB()

# train the model
gnb.fit(X_train, y_train)

# make predictions
y_predict_prob = gnb.predict_proba(X_test)
y_pred = gnb.predict(X_test)

# print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))


precision = precision_score(y_test, y_pred,average='weighted')
print('Precision: %f' % precision)

# Recall
recall = recall_score(y_test, y_pred,average='weighted')
print('Recall: %f' % recall)

# F1 score
f1 = f1_score(y_test, y_pred,average='weighted')
print('F1 score: %f' % f1)
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)

print(report)
```
